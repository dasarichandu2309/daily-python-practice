```python
#Checking the data type of given variables
a = 10
b = 12.5
c = 2+5j
d = "Hello"
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
# Converting between datatypes
a = 10
print(float(a))
print(int(a))
print(str(a))
print(complex(a))
```

```python
#indexing using positive 
a = "HELLO"
print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])
```

```python
#Acsess charecters using negative slicing
a = "HELLOWROLD"
print(a[-10:-5])
print(a[-5:])
print(a[-7:-4])
```

```python
# slicing the string using step value [0:5 :2] the value 2 is the step size
a = "HELLOWORLD"
print(a[0:10:2])
print(a[2:8:2])
```

```python
# Reversing a string using slicing
a = "CHANDU"
b = "CADAC"
print(b[::-1])
print(a[::-1])
```

```python
# Count vowels in a given stirng
a = "HELLOWORLD"
vowel = 'AEIOU'
count=0
for ch in (a):
    if ch in (vowel):
        count+=1
print("The total number of vowels :" , count)
```
