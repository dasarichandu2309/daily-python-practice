## Indexing

```python
a = "HELLO"
print("Positive Indexing :")
print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])
print("Negative Indexing:")
print(a[-1])
print(a[-2])
print(a[-3])
print(a[-4])
print(a[-5])
```

## Range()
## slicing
### positive indexing

```python
a = "HELLO" #H-0,E-1,L-2,L-3,O-4
print(a[0:5])
print(a[1:4])
print(a[:5])
print(a[1:])
print(a[2:])
print(a[:2])
print(a[:])
print(type(a))
```

# Negative indexing

```python
b = "HEADPHONES"
print(b[-9:-4])
print(b[-6:-1])
print(b[-10:-6])
print(b[-4:-1])
print(b[:])
print(b[-10:])
```

# Step Size

```python
b = "HEADPHONES"
print(b[0:10:2])
print(b[0:11:3])
print(b[-10:-1:2])
print(b[::-1])
```
