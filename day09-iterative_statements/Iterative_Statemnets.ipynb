```python
#set
animals = {'cat' , 'dog' , 'horse'}

for i in animals:
    print(i , end =' ')
```

```python
#dictionary
student = {'name':'john' , 'age':20 , 'grade':'a'}
for i in student:
    print(i)
print("----------------------")
for i in student.values():
    print(i)
print("-----------------------")
for keys, values in student.items():
    print(keys ,":" , values)
```

```python
numbers = [2,4,6,8,10]

result = 0
for i in numbers:
    result += i
print(result)
```

```python
numbers = [2,4,6,8,10]

result = 0

for i in numbers:
    result = result + i**2
    print(result)
print("sum of square:" , result)
```

```python
# ennumerate()

names = ["A"  ,"B" ,"C"]
for index,name in enumerate(names):
    print(f"{index}:{name}")
```

```python
#zip()
roll_no = [102,103,104]
students = ['A' , 'B' , 'C']

for r,s in zip(roll_no , students):
    print(f"Roll no: {r}, NAme: {s}")
```

```python
#question

names = ["A" , "B" ,"C"]
marks = [95 , 94 ,97]
for n , s in zip(names,marks):
    print(f"{n} scored {s}")
```

```python
#factorial

number = 10
result = 1

for i in range(1, number+1):
    result*= i
print(f"The factorial of {number} is {result}")
```

## while condition

```python
count = 1
while count <= 5:
    print(count)
    count += 1
    
```

```python
i = 1

while i<= 10:
    print(i , end = " ")
    i+=1
```

```python
i = 1

while i<51:
    if i%5 == 0 or i%7 == 0:
        print(i , end =' ')
    i += 1
        
```

```python
# sum of all even numbers from 1 to 50 using while loop

num = 1
total_sum = 0

while num<= 50:
    if num % 2 == 0:
        total_sum += num
    num+=1
print("Sum of all even numbers from 1 to 50 is:",total_sum)
```

```python
# multiplication table

num  = 5

result = 1
print("The multiplication table of:",num)

while result <= 10:
    s = num * result
    print(num , 'X' , result,'=', s)
    result += 1
```

```python
# Guess the secrect number using while loop
secret_number = 7
guess = 0
while guess != secret_number:
    guess = int(input("Guess the number:"))
    if guess != secret_number:
        print("Try again")
print("Congratulations! you guessed secret number")
```

```python
# Reverse a number using while loop
a = 123456
b = str(a)
c = (b[::-1])
print(c,type(c))
a =int(c)
print(a,type(a))
```

```python
#### Reverse a number using while loop
num = int(input("Enter a number:"))
reverse = 0

while num>0:
    #get the list digit
    digit = num % 10
    reverse = reverse * 10 + digit
    num  = num//10
print("Reversed number:" , reverse)
    
```

## Break statement

```python
for i in range(1,6):
    print(i)
    if i == 3:
        
        break
```

```python
numbers  = [11 , 53, 75, 42, 55]
for i in numbers:
    print(i)
    if i % 2 == 0:
        print("Even number found:" , i)
        break
```

```python

num  = 1
while num < 6:
    print(num)
    if num == 3:
        break
    num+=1
```

```python
numbers  = [11 , 53, 75, 42, 55,48] 

i = 0 # index value
while i < len(numbers):
    num = numbers[i]
    print(num)

    if num % 2 == 0:
        print("Even number found:" , num)
        break
    i+=1
    
```

## Continue Statement

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
```

```python
count = 0

while count < 5:
    count +=1
    if count == 3:
        continue
    print(count)
```
