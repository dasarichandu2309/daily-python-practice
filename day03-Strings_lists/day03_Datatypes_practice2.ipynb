```python
# strip the spaces from a string
a = "     HelloWorld"
b = "HelloWorld     "
print(a.strip())
print(b.strip())
```

```python
# Replace the charecters in string using replace()
a = "A boy is cute"
print(a.replace("boy" , "girl"))
```

```python
# Split the string by commas using slpit()
a = 'cat , dog , pig'
print(a.split(","))
```

```python
# Check weather string starts with given prefix using startswith()
a = "Data science"
print(a.startswith("Data"))
```

```python
# Check weather string ends with given sufix using endswith()
a = "Im a good boy"
print(a.endswith("boy"))
```

```python
# Find the index of given Stirng using find()
a = "Hello"
print(a.find('l'))
```

```python
# Count the times a charecter appers in stirng using count()
a = "Bananaa"
print(a.count("a"))
```

```python
# Check weather a string contains only alphabets using isalpha()
a = "Hello"
print(a.isalpha())
```

```python
# Check weather a stirng contains only digits using isdigit()
a = '1234'
print(a.isdigit())
```

```python
# Check weather a string contains both alphabetes and digits using isalnum()
a = "Hell123"
print(a.isalnum())
```

```python
# Capitalize the first string element using capitalize()
a = "chandu"
print(a.capitalize())
```
