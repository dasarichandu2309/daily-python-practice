```python
#11. Perfect Square Checker
#Problem Statement:
#Write a program to check whether a given number is a perfect square, using a for loop.
n  = int(input("Enter the number:"))
perfect_square = False
for i in range(1 , n+1):
    if i*i  == n:
        perfect_square = True
if perfect_square :
    print("perfect square")
else:
    print("not a perfect square")
```

```python
#Prime Numbers up to N
#Problem Statement:
#Write a program that prints all prime numbers up to a given number N, using a for loop.
num  = int(input("enter the range"))
for num in range(2, num+1):
    is_prime = True
    for i in range(2,num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num , end = " ")
            
```

```python
#3. Fibonacci Series Generator
#Problem Statement:
#Write a program that generates the Fibonacci series up to N terms, using a for loop.
n = int(input("Enter the range:"))
a = 0
b = 1
print(a,b ,end =" ")
for i in range(2,n):
    c = a+b  
    a=b
    b=c
    print(c, end = " ")
    
```

```python
#14. Armstrong Number Checker
#Problem Statement:
#Write a program to check whether a given number is an Armstrong number, using a for loop.
n = input("Enter the number:")
result = 0
for ch in n:
    char = int(ch)
    b= char **3
    result = b+result
if result == int(n):
        print("amstrong")
else:
        print("Not an amstrong")
```

```python
#17. Common Elements Finder
#Problem Statement:
#Write a program that finds and prints common elements between two lists using a for loop.
n = int(input("enter the range:"))
l1 = []
for i in range(n):
    e = input("Enter the the elements:")
    l1.append(e)
n1 = int(input("enter the the range:"))
l2 = []
for i in range(n1):
    f = input("enter the elements :")
    l2.append(f)
res = []
for j in l1:
    if j in l2:
        res.append(j)
print(res)
```

```python
#19. Palindrome Word Checker
#Problem Statement:
#Write a program to check whether a given word is a palindrome, using a for loop.
a = input("Enter the string")
length  = len(a) - 1
rev = ""
for ch in a:
    rev = rev + a[length]
    length-=1
if rev  == a:
    print("palindrome")
else:
    print("not a palindrome")
```
