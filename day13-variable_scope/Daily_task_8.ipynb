```python
#Remove prefix and suffix from string
#Description: Remove specified prefix and suffix, e.g., '##' from '##Data-Science##'.
a = input("enter the string:")
result = ""
for ch in a:
    if ch.isalpha():
        result+=ch
    elif ch == ('-'):
        result+= '-'
print(result)
    
```

```python
#Insert spaces before capital letters except first
#Description: Insert a space before each capital letter in camel-case string except the first
a =input("Enter the string")
result = a[0]
for ch in a[1:]:
    if ch.isupper():
        result+= " " + ch 
    else:
        result+=ch
print(result)
```

```python
#### Insert elements in the middle of a list
#Description: Insert [88,99] in the middle of a list.
a = int(input("enter the range:"))
li = []
for i in range(a):
    e = input("Enter the elements")
    li.append(e)
middle = len(li)//2
li[middle : middle] = [88,89]
print(li)
```

```python
#Merge two lists at specific index
#Description: Merge [1,2,3] & [4,5] at index 1
a = int(input("enter the range 1:"))
li1 = []
for i in range(a):
    e = int(input("Enter the elements"))
    li1.append(e)
b = int(input("enter the range:"))
li2 = []
for i in range(b):
    e = int(input("Enter the elements"))
    li2.append(e)
index = 1
li1[index:index] = li2
print(li1)
```
