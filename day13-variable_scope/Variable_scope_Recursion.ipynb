## local Varibale/Scope

```python
def greet():
    message = "Hello"  # Here , message is local Variable
    print(message)
greet()
#print(message) ---> Error message is created inside a function , so it exists while the function runs
```

```python
def show_name():
    name = "John"
    print("Name is:" , name)
show_name()
```

```python
def calculate_area(radius):
    pi = 3.14159
    area = pi*radius**2
    return area
result = calculate_area(6)
print(result)
```

## Global scope/Global variable

```python
app_name = "flipkart"   #global variable

def show_name():
    print("Application:" , app_name)
show_name()
print("Application:" , app_name)
```

```python
app_name = "flipkart"   #global variable

def show_name():
    global app_name
    app_name = "Amazon"
    print("Application:" , app_name)
    
show_name()
print("App : " , app_name)
```

```python
status = "offline"
def go_online():
    global status
    status = "online"
    print("status:" , status)
go_online()
print("status2:" , status)
```

## Enclosing variable / non-local scope

```python
def outer():
    x = 10   # this vaible is local to outer() and enclosing(non local) for inner()
    def inner():
        print(x)
    inner()
outer()
```

## Recursion

```python
def factorial_iter(n):
    result = 1
    for i in range(1,n+1):
        result*=i
    return result
factorial_iter(5)
```

```python
# with recurson
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n-1)
        
factorial_recursive(5)
```

```python
def rev_str(s):
    return s[::-1]
print(rev_str("hello"))
```

```python
#with recursion
def reverse_string(s):
    if len(s) <= 1:
        return s
    else:
        return s[-1] + reverse_string(s[:-1])
print(reverse_string("Hello"))
```
