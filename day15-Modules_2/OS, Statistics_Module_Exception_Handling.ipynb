## Statistics Module

```python
# --> find measures like mean, median, mode, variance, standard deviation

import statistics as st

data = [1,2,2,3,4,4,4,5]

# Mean(Average): Sum of values / Number of values

mean_value = st.mean(data)
print("Mean:", mean_value)

# Median : 
# 1. Make sure given numbers must be in ascending order
# 2. It is used to return the middle value of a given data --> odd list
# 3. It takes middle two values and add them and divide by 2 of a given data --> even list

median_value = st.median(data)
print("Median:", median_value)

# Mode : Most frequently occurring value
mode_value = st.mode(data)
print("Mode:", mode_value)

# Variance: Sum of Squared differences / (n-1)
variance_value = st.variance(data)
print("Variance:", variance_value)

# Standard Deviation: Square root of Variance
std_value = st.stdev(data)
print("Standard Deviation:", std_value)
```
