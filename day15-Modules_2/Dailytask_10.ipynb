```python
#Why Return?
#Write a function add(a, b) that returns the sum of two numbers. Call it and print the returned value. 
#Why is return used instead of print inside the function?
def add(a,b):
    result = a+b
    return result
sum_ = add(5,6)
print(sum_)
# using return we can call function the and store the vlaues in variables to reuse them
# using print we can only display the result and cant store or reuse them
```

```python
#Return vs. None
#Write a function without a return statement. Call it and check what it prints. Why does it show None?
def add(a,b):
    result = a+b
sum_ = add(5,6)
print(sum_)
# without return the evaluated value will not be shown it will just be stored
```

```python
#Positional Arguments
#Write a function greet(first, last) that prints "Hello, first last!". Call it with different names, maintaining order.
def greet(name,age):
    print(f"Hello {name}  {age} !")
greet('chandu' , 23)
```

```python
#Keyword Arguments
#Using the same greet function, call it using greet(last="Smith", first="John"). 
#What happens if the order changes?
def greet(name , age):
    print(f"Hello your name is {name} and age is {age}")
greet(name = "chandu" , age = 23)
greet(age = 23 , name = "chandu")
```

```python
#Default Values
#Create a function welcome(name, city="Hyderabad") that prints "{name} is from {city}". Call it with and without providing city.
def welcome(name , city = "hyderabad"):
    print(f"{name} is from {city}")

welcome("chandu")
# in default arguments the postions are first (normalvalue,default) if (default ,normal value) it shows error
# the values in deafault arguments are declared in function itself and null values are replaced with default vlaues
```

```python
#Using args
#Write a function multiply(*nums) that multiplies all numbers passed to it. Call it with 2, 3, 4 and print the result.
def multiply(*args):
    total =1
    for i in args:
        total *=i
    return total
multiply(2,3)
```

```python
#**Using kwargs
#Write a function student_info(**details) that prints each key-value pair passed (e.g., name="Anu", age=20).
def student_info(**kwargs):
    for key,values in kwargs.items():
        print(key ,":",values)
student_info(name = "Anu" , age = 20)
```

```python
#Mixing Positional and Keyword Arguments
#Create a function order(item, price, quantity=1) and call it with a mix of positional and keyword arguments.
def order(item , price, quantity = 1):
    print(f"{item} is {price} amount of quantity {quantity}")

order("biryani" , price = 130)
```
