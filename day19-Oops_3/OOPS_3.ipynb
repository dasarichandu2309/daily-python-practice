## Constructor overloading

```python
# Same constructor __init__ ----> behaves dufferntly depending in arguments passes
class person:
    def __init__(self,name,age = None):
        self.name = name
        self.age = age

p1 = person("John")
print(p1.name , p1.age)
p2 = person("James" , 25)
print(p2.name, p2.age)
```

#### Method overloading ---> Same method name  , Differnt parameters
#### Operator Overloading ----> Same operator behaves differnt for differnt data types
#### Constuctor Overloading ---> Same class COnstructor works differnt depending on arguments


## RunTime Polymorphism

```python
# Dynamically Typed (Duck typing)



class dog:
    def speak (self):
        print("Bark")

class cat:
    def speak(self):
        print("Meow")

class duck:
    def speak(self):
        print("Quack")


def make_sound(animal):
    animal.speak()

d = dog()
c = cat()
k = duck()

make_sound(d)
make_sound(c)
make_sound(k)

```

#### Method overriding

```python
class animal:
    def speak(self):
        print("Some sound")

class dog(animal):
    def speak(self):
        print("Bark")


d = dog()
d.speak()
```

```python
## Super() ----> used inside a child class to call the methods from its parent class

class animal:
    def speak(self):
        print("Some sound")

class dog(animal):
    def speak(self):
        super().speak()
        print("Bark")


d = dog()
d.speak()
```

### Constructor Overriding



```python
class parent:
    def __init__(self):
        print("Parent constructor")

class child(parent):
    def __init__(self):
        super().__init__()
        print("Child constructor")

obj = child()
```

## Inheritnace ---> Accessing variables and methods

```python
class parent:
    def __init__(self):
        self.name = "Parent Class"
        self.value = 100

    def show(self):
        print("This is parent class method")
        print("Name:", self.name)
        print("Value" , self.value)

class child(parent):
    def __init__(self):
        super().__init__()
        self.child_name  = "Child class"
    def display(self):
        print("this is child class method")
        print("Name:" , self.child_name)



obj = child()
obj.show()
obj.display()
    
```

## Getter Method and Setter Method


```python
class Student:
    def __init__(self,name,marks):
        self.__name = name  #Private varibale
        self.__marks = marks # Private variable

    #Public method(Getter)
    def get_marks(self):
        return self.__marks
        

    def set_marks(self,new_marks):
        if 0 <= new_marks <= 100:
            self.__marks = new_marks
        else:
            print("Invalid marks! Must be between 0  and 100")

s1 = Student("john" , 82)
print(s1.get_marks())

s1.set_marks(95)
print(s1.get_marks())

s1.set_marks(105)
print(s1.get_marks())

        
```

## Encapsulation

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.__owner = owner
        self.__balance = balance

    def get_owner(self):
        return self.__owner
    def get_balance(self):
        return self.__balance


    def deposit(self,amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount} , New Balance = {self.__balance}")
        else:
            print("Deposited amount must be positive")

    def withdraw(self,amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount} , Remaing Balance = {self.__balance}")
        else:
            print("Invalid Withdrawl amount")

account = BankAccount("John" , 1000)
print(account.get_owner())
print(account.get_balance())


account.deposit(200)
print(account.get_balance())

account.withdraw(300)
print(account.get_balance())
```
