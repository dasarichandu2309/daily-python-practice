 ## Step 1: Lambda Function


```python
#Write a function using lambda that takes a number and returns its square.
a = lambda i : i**2
a(4)
```

```python
#Create a lambda function that takes two numbers and returns their sum.
a = lambda x,y : x+y
a(2,3)
```

```python
#Write a lambda function that checks if a number is even.
a = lambda i : i % 2 == 0
a(4)
```

```python
#Create a lambda function that converts a temperature from Celsius to Fahrenheit.
a = lambda i : i * 9/5 + 32
a(45)
```

```python
#Write a lambda function that takes a string and returns it in uppercase.
a = lambda i : i.upper()
a("chandu")
```

## Step 2: Higher-Order Functions 

```python
#Given a list of numbers, write a program to double each number in the list.
a = [1,2,3,4,5,6,7,8]
b = list(map(lambda i : i*2 , a))
print(b)
```

```python
#Given a list of numbers, write a program to keep only the numbers greater than 10.
a = [10,2,30,20,3,4,5,6,7,80,90]
b = list(filter(lambda i : i>10 , a))
print(b)
```

```python
#Given a list of numbers, write a program to calculate the product of all numbers in the list.
from functools import reduce
a = [1,2,3,4,4,5,5,6,7]
b = reduce(lambda i,j: i*j ,a)
print(b)


```

## Step 3: Higher-Order Functions 



```python
#Given a list of numbers, write a program to triple each number using a single expression.
a = [1,2,3,4,4,5,6,7]
b = list(map(lambda i: i *3 ,a))
print(b)
```

```python
#From a list of numbers, write a program to keep only the odd numbers using a single expression.
a = [1,2,3,4,4,5,6,7]
b = list(filter(lambda i : i%2!=0 , a))
print(b)
```

```python
#From a list of numbers, write a program to calculate the sum of all squares of the numbers.
from functools import reduce
a = [1,2,3,4,4,5,6,7]
b = list(map(lambda i : i**2 , a))
c = reduce(lambda x,y : x+y , b)
print(b)
print(c)
```

```python
#Given a list of strings, write a program to find the length of each string in a single expression.
a = ['chandu' , 'kaif' , 'prasad' , 'vinod', 'sanath' , 'rohith']
b = list(map(lambda i: len(i) , a))
print(b)
```

```python
#From a list of numbers, write a program to find the largest number using a single expression.
from functools import reduce
a = [1,2,3,4,4,5,6,7]
b = reduce(lambda x,y: x if x>y else y , a)
print(b)
```

```python
#Given a list of numbers, write a program to calculate the sum of the cubes of all even numbers.
from functools import reduce
a = [1,2,3,4,4,5,6,7]
b = list(filter(lambda i: i % 2 == 0 ,a))
c = list(map(lambda i : i**3 , b))
d = reduce(lambda x,y : x+y , c)
print(d)
```
