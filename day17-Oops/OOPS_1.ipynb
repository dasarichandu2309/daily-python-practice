### OOPS Example

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def greet(self):
        print(f"HEllo,my name is {self.name} and I am {self.age} years old")

#create object
person1 = Person("Amit" , 20)
person2 = Person("john" , 23)

print(f"{person1.name} is {person1.age} years old")
print(f"{person2.name} is {person2.age} years old")

person1.greet()
person2.greet()

```

```python
class Car:
    def __init__(self,brand,color,fuel):
        self.brand = brand
        self.color = color
        self.fuel  = fuel
    def start(self):
        print(f"The {self.brand} {self.color} with {self.fuel} fuel has started")
    def stop(self):
        print(f"The {self.brand} {self.color} with {self.fuel} fuel has stopped")
car1  = Car("Honda" , "red" , "petrol")
car2  = Car("BMW" , "Blue" , "deisel")

print(car1.brand , car1.color , car1.fuel)
print(car2.brand , car2.color , car2.fuel)

car1.stop()
car1.start()
car2.start()
car2.stop()
```

## Acsess Modifiers

### Public Acsess Modifier




```python
class student:

    school_name = "ABC High School"  #class variable
    def __init__(self,name,marks):# public
        self.name = name
        self.marks = marks

    def display_info(self):
        print(f"StudentName:{self.name} , Grade:{self.marks}")

s1 = student("Ram" , 'A')

print(s1.name, s1.marks)

s1.display_info()
```

### Private Acsess modifier

```python
class student:
    def __init__(self,name,marks):
        self.__name = name#privte variable
        self.__grade = marks# private variable

    def __calculate_result(self):
        if self.__grade == "A":
            return "Excellent"
        elif self.__grade == "B":
            return "Good"
        else:
            return "Needs Improvement"
    def show_result(self):
        print(f"StudentName:{self.__name} , Grade:{self.__grade}")
        print("Result:" , self.__calculate_result())

s1 = student("ravi" , "B")
s1.show_result()
```

## Protected acsess modifier

```python
class student:
    def __init__(self,name,marks):
        self._name = name#protected variable
        self._grade = marks# protected variable
        
    def show_result(self):
        print(f"StudentName:{self._name} , Grade:{self._grade}")

s1 = student("Ram" , "A")
#print(s1._name) ---> Sitll works, but should be avoided
s1.show_result()
```

## Inheritance

```python
class animal:
    def eat(self):
        print("Eat")
    def sleep(self):
        print("Sleep")

# child class

class dog(animal):    # Dog inherits from animal
    def bark(self):
        print("Barks")


class Cat(animal):    # cat inherits from animal
    def meow(self):
        print("Meow")

Dog = dog()
Dog.eat()
Dog.sleep()
Dog.bark()
print("-" * 40)
cat  = Cat()
cat.meow()
cat.eat()
cat.sleep()

```
