# Dictionary methods

```python
student = {'name': "John", 'age': 24 , 'salary':25000}
print(student)

# get() -----> Returns the value for a given key
print(student.get("name"))
print(student.get("country"))
# Default value if key is not present in dictionary
print(student.get("country" , "Not Found"))
```

```python
# update()----> Adds or updates key_value pairs
student = {'name': "John", 'age': 24 , 'salary':25000}
print(student)

student.update({'age': 30})
print(student)

student.update({'age': 30 , 'salary':3000000})
print(student)

# adding item using update()

student.update({"country": "India"})
print(student)
```

```python
#copy() ----> Creates the copy of the dictionary
#del b
#print(b)


b = student.copy()
print(b)
```

```python
# removes a key and return its value
student = {'name': "John", 'age': 24 , 'salary':25000}
print(b)
b_pop = student.pop("age")
print(student)
print(b_pop) # Returns the value of the age
```

```python
#popitem() ---> Removes and returns the last inserted key-value pair

student = {'name': "John", 'age': 24 , 'salary':25000}
print(student)
s_popitem = student.popitem()
print(student)
print(s_popitem)
```

```python
a = ['a','b','c']   # Keys
b = ['apple' , 'baanana', 'carrot']   #Values

d = dict(zip(a,b))
print(d)
```

```python
# enumerate ----> Dictionary wiht index as key

a = ['apple' , 'baanana', 'carrot']
d1 = dict(enumerate(a))
print(d1)
```

```python
student = {'name': "John", 'age': 24 , 'salary':25000}
print(student)

student.clear()
print(student)
```

# Set Operations

```python
# union
a = {1,2,3}
b = {2,3,4,5}

print("Union of sets:",a|b)
print("Union of sets:",a.union(b))
```

```python
# intersection
a = {1,2,3}
b = {2,3,4,5}

print("intersection of sets:",a&b)
print("Intersection of sets:",a.intersection(b))

#intersection_update
a.intersection_update(b)
print(a)
```

```python
# difference

a = {1,2,3}
b = {2,3,4,5}
print("difference between two sets:",a-b)
print("difference between two sets:",b-a)

print("difference between two sets with methods:",a.difference(b))
print("difference between two sets with methods:",b.difference(a))
a.difference_update(b)
print(a)
```

```python
#symmetric differnce
a = {1,2,3}
b = {2,3,4,5}
print("symmetric difference between two sets:",a^b)
print("symmetric difference between two sets with methods:",a.symmetric_difference(b))
```
