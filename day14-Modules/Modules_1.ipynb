```python
# prime number
def is_prime(n):
    if n<=1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(29))
```

## Random module

```python
#random() ----> Ii gives float number that ranges from 0.0 to 1.0

import random
num = random.random()
print(num)
```

```python
# randint() ---> generates integer values from the range of numbers given

num = random.randint(1,10)
print(num)
```

```python
#randrange() ----> selects an item or value from the given range(start:stop:step size)

num  = random.randrange(1,10,2)
print(num)
```

```python
#random.choice() ----> selects or make choice an item from a series at random
s = random.choice("Chandu")
print(s)

s1 = random.choice([1,2,3,4,5])
print(s1)
```

```python
# random.shufflec----> this function shuffles randomly from the given list
a = [1,2,3,4,5,6]
random.shuffle(a)
print(a)
```

```python
from datetime import datetime , date , time , timedelta

now  = datetime.now()  #now() gives current date and time
print("Current date and time:" , now)

today = date.today()  #today() returns curent date/ today date
print("Today's date:" , today)

some_date = date(2023,4,10)
print(some_date)
print(type(some_date))

# formatting date as string
print("formatted date:")
a = now.strftime("%Y-%m-%d %H:%M:%S")
print(a)
print(type(a))
#format string to date time
dt = datetime.strptime("2025-08-11 14:30:10" ,"%Y-%m-%d %H:%M:%S" )
print(dt)

tomorrow  = today + timedelta(days=1)
print(tomorrow)
```
