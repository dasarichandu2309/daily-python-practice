# List Operations/ Methods

```python
# Repetetion
a = [1,2,3,4,5,6]
print(type(a))
print(a*2)
```

```python
# Comcatenation
a = [1,2,3,4,5,6]
b = [2,3,4,5,6,7]
print(a+b)
```

```python
# append() ---> Adds an element to the end of the list
a = [1,2,3]
a.append(4) # Adds only single element
print(a)
```

```python
# extend() ---> Appends the elements of another list to current list
a = [1,2,3]
b = [4,5,6]
a.extend(b)
print(a)
```

```python
a = [1,2,3]
b = [4,5,6]
b.extend(a)
print(b)
```

```python
# insert() ---> Inserts an element at a specified index
a = [1,2,3]
a.insert(1,1.5)  # Insert(index_position,new_value)
print(a)
```

```python
# remove() ---> Removes the first occurence of specified index
a = [1,2,3,2,2,2]
a.remove(2)
print(a)
```

```python
#index() ---> Returns the index of the first occurence of specified element
a = [1,2,3]
b = a.index(3)
print("Index position of 3 :",b)
```

```python
# count() ---> Returns the no.of occurences of a specified element in the list
a = [1,2,3,2,2,2]
b = a.count(2)
print(b)
```

```python
# sort() ---> Sorts the list in ascending order
a = [1,4,5,3,2,6]
a.sort()
print(a)
```

```python
# reverse() ---> REverse the order of elements in the list
print(a)
a.reverse()
print(a)
```

```python
# copy() ---> It creates the copy of the list
a = [1,2,3,4,5]
b = a.copy()
print(b)
```

```python
# clear() ---> removes all the elements from the list
a = [1,2,3,4,5]
a.clear()
print(a)
#a.append(3)
#print(a)
```

```python
# min() ---> Returns the smallest elemnet in the list
# max() ---> Retruns the largest element in the list
a = [45,78,54,98,25]
b = min(a)
c = max(a)
print("min value:",b)
print("max value:",c)
```

```python
# pop() ---> Removes and returns that elemnts
a = [1,2,3,4,5]
b = a.pop(1)
print(a)
print([b])
```

### Deleting list

```python
a = [10,20,30,40,50]
print(a)


# remove()
a.remove(30)
print(a)

#remove by index
a.pop(0)
print(a)

# delete by index
del a[1]
print(a)

# clear all elements
a.clear()
print(a)

# delete whole list data
del a
print(a)
```

```python
# Indexing()
a = (1,1.5,1+3j,"hello")
print(a[0])
print(a[1])
print(a[2])
print(a[3])
# negative indexing
print(a[-1])
print(a[-2])
print(a[-3])
print(a[-4])
```

```python
# slicing
a = (10,20,30,40,50,60,70,)
print(a[0:4])
print(a[0:7:2])
print(a[-6:-2])
print(a[::-1])
```

```python
# Delete a tuple
a = (1,2,3,4,5)
del a
print a
```

```python
# delete individual element from tuple by converting the list
a = (1,2,3,4,5)
b = list(a)
print(b)
b.remove(3)
print(b)
c = tuple(b)
print(c)
```

```python
# Immutable ---> we cannot change elements forming
# a = (1,2,3,4,5,6)
# a[2] = "hello"
```

```python
# Nested tuple ---> tuple inside tuple
a = (1,2,"hello",(3,4,5),("apple","mango"),[1,2,3])
print(a[3])
print(a[3][1])
print(a[5])
a[5][1] = "hello"
print(a)
print(type(a))
```
