## Using list

```python
#Create a dictionary from a list [1, 2, 3, 4, 5] where each element is the key and its cube is the value.
a = [1,2,3,4,5]
b = {i:i**3 for i in a}
print(b)
```

```python
#From a list of words ["apple", "banana", "cherry"], generate a dictionary with words as keys and their lengths as values.
a = ["apple", "banana", "cherry"]
b = {ch:len(ch) for ch in a}
print(b)
```

```python
#Given a list [10, 20, 30, 40], create a dictionary with index as key and element as value.
a = [10, 20, 30, 40]
b={i:a[i]for i in range(len(a))}
print(b)
```

```python
#From a list [2, 4, 6, 8], generate a dictionary with numbers as keys and "Even" as values.
a = [2, 4, 6, 8]
b = {i:"even"for i in range(len(a))}
print(b)
```

```python
#Given a list of names ["Alice", "Bob", "Charlie"], create a dictionary with names as keys and the first letter as value.
names = ["Alice", "Bob", "Charlie"]
b = {ch:ch[0] for ch in names}
print(b)
```

##  Using Dictionaries


```python
#Given a dictionary {"a": 1, "b": 2, "c": 3}, create a new dictionary with keys same but values squared.
dicti = {"a": 1, "b": 2, "c": 3}
new = {keys:values**2 for keys,values in dicti.items()}
print(new)
```

```python
#From a dictionary {"cat": 4, "spider": 8}, generate a dictionary with animal name as key and "Legs: X" as value.
d = {"cat": 4, "spider": 8}
c = {k:"legs:"+ str(v) for k,v in d.items()}
print(c)
```

```python
#Create a dictionary from {"x": 5, "y": 10, "z": 15} where only items with values greater than 7 are included.
d = {"x": 5, "y": 10, "z": 15}
b = {k:v for k,v in d.items() if v>7}
print(b)
```

```python
#From a dictionary {"math": 90, "science": 80, "history": 70}, generate a dictionary where subjects are keys and "Pass"/"Fail" is value (pass â‰¥ 75).
a={"math": 90, "science": 80, "history": 70}
c = {k:"Pass"if v>=75 else "Fail" for k,v in a.items()}
print(c)
```

```python
#Given a dictionary {"pen": 20, "book": 50, "eraser": 5}, create a dictionary with items as keys and price+10 as value.
a = {"pen": 20, "book": 50, "eraser": 5}
b = {k:v+10 for k,v in a.items()}
print(b)
```

## Using Sets


```python
#Convert a set {1, 2, 3, 4} into a dictionary with element as key and its factorial as value.
import math
s = {1, 2, 3, 4}
r = {x: math.factorial(x) for x in s}
print(r)
```

```python
#From a set {"red", "green", "blue"}, create a dictionary with color as key and length of color name as value.
s = {"red", "green", "blue"}
b = {ch:len(ch) for ch in s}
print(b)
```

```python
#Given two sets {1, 2, 3} and {10, 20, 30}, create a dictionary mapping first set elements to second set elements.
s = {1, 2, 3}
s1 = {10, 20, 30}
result = {k: v for k,v in zip(s, s1)}
print(result)
```

```python
#From a set of numbers {2, 3, 5, 7}, generate a dictionary with prime number as key and "Prime" as value.
s = {2, 3, 5, 7}
b = {i:"prime" for i in s}
print(b)
```

```python
#From a set {"car", "bike", "bus"}, create a dictionary with vehicle as key and uppercase as value.
s = {"car", "bike", "bus"}
b = {i:i.upper() for i in s}
print(b)
```
