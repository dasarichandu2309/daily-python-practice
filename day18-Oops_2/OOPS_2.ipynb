## Types of  Inheritance

```python
## single inheritence

class Father:
    def driving(self):
        print("Driving")
class Son(Father):
    def study(self):
        print("Study")

s = Son()
s.driving()
s.study()
```

```python
# multiple Inhertiance

class Father:
    def driving(self):
        print("Driving")
class Mother:
    def cooking(self):
        print("Cooking")
class Child(Father,Mother):
    def study(self):
        print("Study")

c = Child()
c.driving()
c.cooking()
c.study()
```

```python
# Multi level inheritance

class GrandFather:
    def farming(self):
        print("Farming")
class Father(GrandFather):
    def driving(self):
        print("Driving")
class Son(Father):
    def study(self):
        print("Study")

s = Son()
s.farming()
s.driving()
s.study()
```

```python
# Hierarchial Inheritance:

class Parent:
    def family_name(self):
        print("Family name")

class Son(Parent):
    def play_cricket(self):
        print("Playing Cricket")

class Dauther(Parent):
    def dance(self):
        print("Dancing")

s = Son()
s.family_name()
s.play_cricket()

d = Dauther()
d.family_name()
d.dance()
```

## Hybrid Inheritance

```python
class GrandFather:
    def farming(self):
        print("Farming")

class Father(GrandFather):
    def driving(self):
        print("Driving")

class Mother:
    def cooking(self):
        print("Cooking")

class Son(Father , Mother):
    def study(self):
        print("Study")


s = Son()
s.farming()
s.driving()
s.cooking()
s.study()



```

## Abstraction

```python
from abc import ABC,abstractmethod   # ABC ---> Abstract base class

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    def stop(self):
        print("Veichle Stopped")

class Car(Vehicle):
    def start(self):
        print("Car started with key")

class Bike(Vehicle):
    def start(self):
        print("Bike started with self start")

c = Car()
c.start()
c.stop()
print("-" * 40)
b = Bike()
b.start()
b.stop()
    
```

## Polymorphism

```python
class Payment:
    def pay(self):
        pass

class Creditcard(Payment):
    def pay(self):
        print("Paying with credit card")

class Paypal(Payment):
    def pay(self):
        print("Paying using paypal")

class Cash(Payment):
    def pay(self):
        print("Paying using Cash")

methods  = [Creditcard() , Paypal() , Cash()]

for i in methods:
    i.pay()
```

```python
class Calculator:
    def add(self,a,b):
        return a+b
    def add(self , a,b,c):
        return a+b+c
    def add(self , a, b ,c ,d):
        return a+b+c+d


b = Calculator()
b.add(2,3)
```

```python
class Calculator:
    def add(self,*args):
        return sum(args)


b = Calculator()
b.add(2,3)
```

## Operator overloading

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __add__(self,other):
        return Point(self.x + other.x , self.y + other.y)

p1 = Point(1,2)
p2 = Point(3,4)

p3 = p1+p2
print(p3.x , p3.y)
```

```python
class Point:
    def __init__(self,x):
        self.x = x

    def __add__(self , other):
        return Point(self.x + other.x)

p1 = Point("hello")

p2  = Point("World")

p3  = p1+p2


print(p3.x)
```
