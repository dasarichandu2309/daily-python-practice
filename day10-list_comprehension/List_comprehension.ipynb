# List comprehension

```python
# without list comprehension

# generate a list a squres from 1 to 5

squares = []
for i in range(1,6):
    squares.append(i**2)
print(squares)
```

```python
for i in range(1,6):
    print(i**2)
```

```python
# with list comprehension
squares = [i**2 for i in range(1,6)]
print(squares)
```

```python
# generate a list with even number:
even = []
for i in range(1,11):
    if i%2 == 0:
        even.append(i)
print(even)
        
```

```python
even = [i for i in range(1,11) if i%2 == 0]
print(even)
```

```python
# list comprehension
# convert all strings to uppercase
names = ['apple' , 'mango', 'banana']
new  = [i.upper() for i in names]
print(new)
```

```python
pairs = [(i, i**2) for i in range(1,6)]
print(pairs)
```

```python
# Nested list comprehension

matrix = [[i*j for j in range(1,4)] for i in range (1,4)]
print(matrix)
```

```python
# create a list of all numbers from 1 to 50 that are divisible by both 3 and 5
l = [i for i in range(1,51) if i%3==0 and i%5==0]
print(l)
```

# Dictionary 

```python
# create a dictiionary of squares

squares = { i:i**2 for i in range (1,6)}
print(squares)
```

```python
# convert list to dictionary with length
#{name:lenght}

names = ['apple' , 'mango', 'banana']
name_length = {name:len(name) for name in names}
print(name_length)
```

```python
#filter dictionary to include only even values

original = {1:2,2:3,3:4,5:6,7:9}
even_values = {k:v for k,v in original.items() if v%2 ==0}
print(even_values)
```

```python
# swap keys and values
original = {'a':1 ,'b':2, 'c':3}

swapped = {v:k for k,v in original.items()}
print(swapped)
```
