```python
def add(a,b):
    result = a+b
    return result
sum_value = add(5,7)
print(sum_value)
```

```python
def calculate_stats(numbers):
    total = sum(numbers)
    average = total/len(numbers)
    return total,average

data = [10,20,30,40,50,60,70,80]
#call function
a,b = calculate_stats(data)
print("Total:" , a)
print("Average:", b)
```

```python
def divide(a,b):
    if b==0:
        return "cannot divide by zero"
    return a/b
result = divide(10,2)
print(result)
result2 = divide(5,0)
print(result2)
```

```python
def square(x):
    return x**2
def cube(x):
    return(x**3)
def square_cube(x):
    squared = square(x)
    cubed = cube(x)
    return squared , cubed
result = square_cube(3)
print(result)
```

```python
# factorial (5) --> 5*4*3*2*1 ---> 120
def factorial(n):
    result = 1
    for i in range(1,n+1):
        result *=i
    return result
result = factorial(5)
print(result)
```

```python
# Area of rectagle --> length * breadth(width)

def area_of_rectangle(length,width):
    result = length*width
    return result
length = 3
width = 5
area = area_of_rectangle(length,width)
print(f"area of rectangle with length {length} and width {width} : {area}")
```

```python
# even odd checker
def check_even_odd(n):
    if n % 2 == 0:
        return "Even number"
    else:
        return "odd  number"
n = int(input("Enter the number:"))
print(check_even_odd(n))
```

```python
# check palindrome
def is_palindrome(word):
    return word==word[::-1]
print(is_palindrome('madam'))
print(is_palindrome("hello"))
```

```python
# fibonnaci
def fibonnaci(n):
    a,b = 0,1
    for i in range(n):
        print(a,end= " ")
        a,b = b, a+b
fibonnaci(10)
```

```python
# check leap year

def leap_year(year):
    if (year%4 == 0) and year %100 !=0 or year %400 == 0:
        return True
    else: return False
l = int(input("Enter the yeear:"))
print(leap_year(l))
```
